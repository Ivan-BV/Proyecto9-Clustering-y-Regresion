```python
%reload_ext autoreload
%autoreload 2
import pandas as pd
import sys
sys.path.append("../")
import src.soporte_ajuste_regresion as sar

import warnings
warnings.simplefilter("ignore")
```

# Cluster 0

```python
data_pickle_cluster0 = pd.read_pickle("../datos/output/df_cluster0.pickle")
df_cluster0 = pd.DataFrame(data_pickle_cluster0)
df_cluster0
```

```python
df_cluster0["clusters_kmeans"] = df_cluster0["clusters_kmeans"].astype(int)
df_cluster0.dtypes
```

```python
lista_modelos = ["regresion", "tree", "random_forest", "gradient_boosting", "xgboost"]
df_metricas_cluster0 = pd.DataFrame()
soporte = sar.AnalisisModelosRegresion(df_cluster0, "Sales")
for modelo in lista_modelos:
    print(f"------------------ Uilizando {modelo.upper()} ------------------")
    print("----------------------------------------------------------------")
    soporte.ajustar_modelo(modelo)
    df_resultados = soporte.obtener_resultados(modelo)
    df_calculo = soporte.calcular_metricas(modelo)
    df_metricas_cluster0 = pd.concat([df_metricas_cluster0, df_calculo], axis=1)
    if modelo is not "regresion":
        soporte.importancia_predictores(modelo)
    
display(df_resultados)
df_metricas_cluster0.head().T
```

```python
df_resultados_colores = df_metricas_cluster0.T.copy()
df_resultados_colores.reset_index(inplace = True, drop=True)
df_resultados_colores = df_resultados_colores.style.apply(sar.color_filas_por_modelo, axis=1)
lista_labels = []
for i in range(0, 10):
    if i % 2 == 0:
        lista_labels.append("test")
    else:
        lista_labels.append("train")
df_resultados_colores.relabel_index(lista_labels)
df_resultados_colores
```
